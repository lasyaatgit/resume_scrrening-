```python
from pdf2image import convert_from_path
import pytesseract

def ocr_pdf_to_text(pdf_path):
    images = convert_from_path(pdf_path)
    text = ""
    for img in images:
        text += pytesseract.image_to_string(img)
    return text

resume_text = ocr_pdf_to_text(pdf_path)
print("Extracted Resume Text (OCR):\n", resume_text[:1000])
```

```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'\d+', '', text)

    text = re.sub(r'[^\w\s]', '', text)
    text = " ".join([word for word in text.split() if word not in stopwords.words('english')])
    return text

clean_resume = clean_text(resume_text)
print("Cleaned Resume:\n", clean_resume[:500])
```

```python
job_description = """
We are looking for a Data Scientist skilled in Python, Machine Learning, Deep Learning, and NLP.
Candidates should have experience with TensorFlow, PyTorch, and SQL.
"""
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([clean_resume, job_description])

match_score = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])
print("\nResume Match Score:", round(match_score[0][0] * 100, 2), "%")

```

```python
REQUIRED_KEYWORDS = {"python", "machine learning", "sql", "deep learning", "nlp"}
```

```python
def keyword_match(resume_text, required_keywords):
    words = set(resume_text.lower().split())
    matched_keywords = words.intersection(required_keywords)
    match_percentage = (len(matched_keywords) / len(required_keywords)) * 100
    return matched_keywords, match_percentage
```

```python
matched_keywords, keyword_match_score = keyword_match(clean_resume, REQUIRED_KEYWORDS)

print(f"Matched Keywords: {matched_keywords}")
print(f"Keyword Match Score: {round(keyword_match_score, 2)}%")
KEYWORD_THRESHOLD = 50
if keyword_match_score < KEYWORD_THRESHOLD:
    print("\nShortlisting Decision: No (Missing Key Skills)")
else:
    print("\nShortlisting Decision: Yes (Meets Skill Requirements)")
```
